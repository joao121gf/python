```python
import pandas as pd
from IPython.display import display
df = pd.read_csv('supermarket_sales - Sheet1.csv', sep=',')
display(df.columns)
```

```python
display(df.head(2))
display(df.shape)
```

```python
# FILTERING INFORMATIONS
#Take all sales by 'Yangon' city



yagon_sales = df[df['City'] == 'Yangon'][['City', 'Invoice ID']]
yagon2 = df[df['City'] == 'Yangon']
display(yagon_sales)
display(yagon2)



```

```python
# Fazendo uma mÃ©dia

# Para cada tipo de produto, descubra qual Ã© o preÃ§o mÃ©dio que foi cobrado.

media_price = df.groupby('Product line')['Unit price'].mean().sort_values(ascending=False)
media_price = df.groupby('Product line')['Unit price'].mean().sort_values(ascending=True)

display(media_price)
```

```python
# Descobrindo o dia da semana

# Olhe a data das compras e crie uma nova informaÃ§Ã£o dizendo se foi feita 
# numa segunda, terÃ§a, quarta, etc.

# Convertendo coluna de data para datatime
df['Date'] = pd.to_datetime(df['Date'])
df['Day of Week'] = df['Date'].dt.day_name()
display(df.head(2))

# descobrindo quais melhores dias da semana para vendas

best_days_week = df.groupby('Day of Week')['Total'].sum().sort_values(ascending=False)
display(best_days_week)
```

```python
# QUAL FILIAL VENDEU MAIS NO TOTAL
best_branch = df.groupby('Branch')['Total'].sum().sort_values(ascending=False)
display(best_branch)
```

```python
# Quais tipos de pagamento os Clientes 'Normal' e 'Member' mais utilizam
method_payment = df.groupby(['Customer type', 'Payment']).size().unstack(fill_value=0)

# Exibir os resultados
display(method_payment)
```

```python
# Clientes que compraram mais de uma vez
more_1 = df.groupby('Invoice ID')['Invoice ID'].count()
display(more_1)
```

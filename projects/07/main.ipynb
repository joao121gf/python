```python
import pandas as pd
from IPython.display import display

# Criando um DataFrame de exemplo
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela'],
    'Idade': [25, 30, 28, 35],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba'],
    'Salário': [3500, 4200, 3900, 4500]
}

df = pd.DataFrame(dados)
display(df)
```

```python
df['Desconto'] = df['Salário'].apply(lambda x: x * 0.95)
display(df)
```

```python
def takeAge(age):
    return 'Mais que 30' if age > 30 else 'Menos que 30'
df['Categoria de idade'] = df['Idade'].apply(takeAge)
display(df)
```

```python
df['Resumo'] = df.apply(lambda linha: linha['Nome'] + ' Mora em ' + linha['Cidade'], axis=1)
display(df)
```

```python
data2 = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
    'Idade': [25, 30, 28, 35, 40],
    'Salário': [3500, 4200, 3900, 4500, 5000],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba', 'São Paulo'],
    'Vendas': [10, 15, 12, 18, 20],
    'Meta': [12, 14, 15, 16, 18]
}

df2 = pd.DataFrame(data2)
display(df2)

```

```python
# Crie uma nova coluna chamada Idade_em_5_anos que contém a idade + 5 anos.
# (Dica: use apply() para adicionar 5 a cada valor da coluna "Idade")

df2['Idade_em_5_anos'] = df2['Idade'].apply(lambda i: i + 5)
display(df2)
```

```python
# Crie uma nova coluna Bateu_Meta, onde o valor seja "Sim" se Vendas for maior ou igual a Meta, e "Não" caso contrário.

def calculate(vendas, meta):
    return 'Sim' if vendas > meta else 'Não'
df2['Bateu_meta'] = df2.apply(lambda row: calculate(row['Vendas'],row['Meta']), axis=1)
display(df2)
```

```python
# Crie uma nova coluna Faixa_Salarial com os seguintes valores:

# "Baixa" para salários até 4000
# "Média" para salários entre 4001 e 4500
# "Alta" para salários acima de 4500


valores = [
    'Baixa', 'Média', 'Alta'
]

def salaryy(value):
    if value < 4000:
        return valores[0] 
    elif value > 4000 and value < 4500:
        return valores[1] 
    else:
        return valores[2]
df2['Faixa_Salarial'] = df2.apply(lambda row: salaryy(row['Salário']), axis=1)
display(df2)
```

```python
df2['Nome_Tamanho'] = df2['Nome'].apply(lambda x: len(x))
display(df2)
```

```python
desemp = [
    'Excelente',
    'Bom',
    'Precisa Melhorar'
]
def funcc(vendas, meta):
    if vendas > meta * 1.2:
        return desemp[0]
    elif vendas >= meta:
        return desemp[1]
    else:
        return desemp[2]



df2['Desempenho'] =  df2.apply(lambda row: funcc(row['Vendas'], row['Meta']), axis=1)
display(df2)
```

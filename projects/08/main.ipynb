```python
import pandas as pd
from IPython.display import display

# Criando um DataFrame de exemplo
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela'],
    'Idade': [25, 30, 28, 35],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba'],
    'Salário': [3500, 4200, 3900, 4500]
}

df = pd.DataFrame(dados)
display(df)
```

```python
df['Idade Dobrada'] = df['Idade'].map(lambda x: x* 2)
display(df)
```

```python
mapeamento_cidades = {
    'São Paulo': 'SP',
    'Rio de Janeiro': 'RJ',
    'Belo Horizonte': 'BH',
    'Curitiba': 'PR'
}

df['Sigla cidade'] = df['Cidade'].map(mapeamento_cidades)
display(df)
```

```python
# 1️⃣ Multiplique todos os salários por 1.05 (aumento de 5%).
# 2️⃣ Use um dicionário para substituir os nomes:

# "Ana" → "Ana Souza"
# "Bruno" → "Bruno Lima"
# "Carlos" → "Carlos Mendes"
# "Daniela" → "Daniela Costa"
# 3️⃣ Crie uma coluna "Imposto", onde cada pessoa paga 12% do salário.
```

```python
nomes = {
    'Ana' : 'Ana Souza',
    'Bruno': 'Bruno Lima',
    'Carlos': 'Carlos Mendes',
    'Daniela': 'Daniela Costa'
}

df['Salário'] = df['Salário'].map(lambda x: x * 1.05)
df['Nome Completo'] = df['Nome'].map(nomes)
df['Imposto'] = df['Salário'].map(lambda x: x * 0.12)

display(df)
```
